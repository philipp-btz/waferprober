```python
import custom_analyzer as canal
import matplotlib.pyplot as plt
import os


folder = f"figs/"
filename = "single_measurement_error.png"

if not os.path.isdir(str(folder)):
        os.makedirs(str(folder))

fig = plt.figure(figsize=[16, 9])


files = canal.get_all_filenames(startdir= "data_storage/Wafer_at_2025_06_26-10_50_56")
dataset = canal.Dataset()
dataset.add_files(files=files, only_measurements=True)#
slicename, slice = dataset.filter(parameter="tag_3", value="single_measurement_error")
dataset.clean_slice_data_rbias(target_slice_name=slicename, verbose=True)
xy_data = dataset.get_slice_mean(target_slice_name=slicename, rbias_only=True)
x = []
y = []  
for key, value in xy_data.items():
    x.append(key)
    y.append(value)

err_data = dataset.get_slice_se(target_slice_name=slicename, rbias_only=True)
err_y = []
for key, value in err_data.items():
    err_y.append(value)
    print(f"Voltage: {key}, error: {value}")


plt.errorbar(x, y, yerr=err_y, fmt=":", capsize=5, label="01E20", capthick=0.5, markevery=1, ecolor="red", barsabove=True)

plt.legend()
plt.legend(fontsize=10, loc="best", ncols=8)#, bbox_to_anchor=(0.6, -0.1))

plt.title(f"single_measurement_error", fontsize=20)
plt.yscale("log")
plt.xscale("linear")
#plt.xlim(left=40, right=48)
plt.axhline(0, color='black')
plt.axvline(0, color='black')   
plt.xlabel("Voltage (V)")
plt.ylabel("Current (A)")
plt.grid(True)
fig.set_facecolor('w')
fig.savefig(os.path.join(folder, filename) if folder else filename, bbox_inches='tight', dpi=500)
plt.close(fig)
print(f"Graph saved to {os.path.join(folder, filename) if folder else filename}")
```
