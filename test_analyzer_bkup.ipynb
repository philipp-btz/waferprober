```python
### print ALL MEASUREMENTS
## (reworked)

import custom_analyzer as canal
import time
import matplotlib.pyplot as plt
import custom_pathlibrary as cpath
import numpy as np
import math
import json
import os
files = canal.get_all_filenames(startdir= "data_storage/Wafer_at_2025_06_16-18_35_06")

for i in files:
    try:
        if not os.path.isfile(i):
            raise FileNotFoundError(f"File {i} does not exist.")
        if i.endswith(".json") and "IV_" in i:
            with open(i, 'r', encoding='utf-8') as file:
                data = json.load(file)
            identifier = data["identifier"]
            if "02" in identifier:
                rbias_temp = data["iv_data_rbias"]
                rbias_data = {}
                for p, q in rbias_temp:
                    rbias_data[p] = q
                i2 = i.replace("/", "+")
                filename = f"{identifier}_{data['die_nr']}_{data['diode_Nr']}_{data['time_at_start']}_{i2}.json"
                graph = canal.Graph(yscale="log", title=f"{filename}")
                graph.unpack_dict(data=rbias_data, mode= "x_y")
        
                graph.save(folder=f"figs/ALL/{identifier}")
    except Exception as e:
        print(f"Error processing file {i}: {e}")
        continue
```
